Решить уравнение:

a) ${y}'-{y}=2x-3$


Для решения этого линейного дифф.уравнения первого порядка используем метод интегрирующего множителя.

Найдём интегрирующий множитель. Уравнение вида {y}' + p(x){y} = g(x) имеет интегрирующий множитель, равный exp(∫p(x)dx).

В данном случае p(x) = -1, поэтому интегрирующий множитель равен exp(∫-1dx) = exp(-x).

Умножим обе части уравнения на интегрирующий множитель. Получим:
exp(-x){y}' - exp(-x)y = (2x-3)exp(-x)

Применим правило производной произведения для левой части уравнения:
(exp(-x){y})' = (2x-3)exp(-x)

Проинтегрируем обе части уравнения по x:
exp(-x){y} = -2xexp(-x) + 3exp(-x) + C, где C - произвольная постоянная.

Запишем уравнение относительно y:
{y} = -2x + 3 + Ce^(x), где C - произвольная постоянная.

Т.о решение данного дифф.уравнения имеет вид: {y} = -2x + 3 + Ce^(x), где C - произвольная постоянная.









Решить уравнение:

b) $x^2{y}'+xy+1=0$


Это неоднородное дифф.уравнение первого порядка, не подходящее под типичные уравнения вида {y}' + p(x){y} = g(x), поэтому для его решения используем другой метод.

Заметим, что уравнение имеет вид:

x^2{y}' + x{y} = -1

Перепишем его в виде производной от произведения функций:

(x^2{y})' = -1

Интегрируем обе стороны по x:

x^2{y} = -x + C

Разрешаем уравнение относительно y:

{y} = (-1/x) + C/x^2

Таким образом, решение данного дифференциального уравнения имеет вид {y} = (-1/x) + C/x^2, где C - произвольная постоянная.

2.Исследовать ряд на сходимость:

  $\sum_{n=1}^{\infty }=\frac{n+2}{n^2+n+1}$

РЕШЕНИЕ:

Можно воспользоваться признаками сравнения.

Для любого $n\in\mathbb{N}$ выполняется неравенство:

$\frac{n+2}{n^2+n+1}\leq\frac{n+2}{n^2}$=
$\frac{1}{n}+\frac{2}{n^2}$
 
Так как ряд $\sum_{n=1}^{\infty}\frac{1}{n}$ расходится,
а ряд $\sum_{n=1}^{\infty}\frac{2}{n^2}$ сходится (это ряд сходится как ряд с постоянными знакопеременными членами), то из признака сравнения следует, что исходный ряд $\sum_{n=1}^{\infty}\frac{n+2}{n^2+n+1}$ расходится.

Таким образом, данный ряд расходится.

Исследовать ряд на сходимость:


$\sum_{n=1}^{\infty }\frac{n^2}{n!}$

РЕШЕНИЕ:

Для исследования сходимости данного ряда можно воспользоваться признаком Даламбера.

Пусть $a_n = \frac{n^2}{n!}$.

Тогда:
$\frac{a_{n+1}}{a_n}=\frac{(n+1)^2}{(n+1)!}\cdot\frac{n!}{n^2}=\frac{n+1}{n^2}$ 
.

Для доказательства сходимости необходимо установить, что ряд

$\sum_{n=1}^{\infty }\frac{n+1}{n^2}$

сходится. Исследуем его с помощью интегрального признака:

$\int_1^\infty \frac{x+1}{x^2}\mathrm{d}x=\int_1^\infty \frac{1}{x}\mathrm{d}x+\int_1^\infty \frac{1}{x^2}\mathrm{d}x<\infty$,



следовательно, данный ряд сходится,

а ряд:

$\sum_{n=1}^{\infty }\frac{n^2}{n!}$ также сходится по признаку Даламбера.

Ответ: ряд $\sum_{n=1}^{\infty }\frac{n^2}{n!}$ сходится.



Исследовать ряд на сходимость:


$\sum_{n=1}^{\infty }n\cdot \left(1-\frac{1}{n} \right)^{n^{2}}$

РЕШЕНИЕ:

Для исследования сходимости данного ряда можно воспользоваться признаком Коши.

Пусть

$a_n = n\cdot \left(1-\frac{1}{n} \right)^{n^{2}}$. 

Тогда

$\sqrt[n]{a_n} = \left(1-\frac{1}{n} \right)^{n}\rightarrow \frac{1}{e}$ 


при $n\rightarrow \infty$

это следует из известного предела
$\lim_{n\rightarrow \infty }\left(1+\frac{1}{n}\right)^{n}=e$
и того факта, что
$\left(1-\frac{1}{n} \right)^{n}\leqslant\frac{1}{e}$ для всех $n$.

Таким образом,

ряд $\sum_{n=1}^{\infty }a_n$ сходится при $\frac{1}{e}<1$ и расходится при $\frac{1}{e}>1$.

Следовательно, исходный ряд $\sum_{n=1}^{\infty }n\cdot \left(1-\frac{1}{n} \right)^{n^{2}}$ сходится при $\frac{1}{e}<1$ и расходится при $\frac{1}{e}>1$.

Ответ: ряд $\sum_{n=1}^{\infty }n\cdot \left(1-\frac{1}{n} \right)^{n^{2}}$ сходится при $\frac{1}{e}<1$ и расходится при $\frac{1}{e}>1$.





Исследовать ряд на сходимость:


$\sum_{n=1}^{\infty }\frac{(-1)^{n+1}}{2n-\ln{n}}$

РЕШЕНИЕ:


Для исследования сходимости данного ряда воспользуемся признаком Лейбница.

Пусть $a_n = \frac{1}{2n-\ln{n}}$.

Тогда: $a_n>0$ для всех $n$;

$a_n$ монотонно убывает к нулю, так как


$a_{n+1}-a_n = \frac{1}{2(n+1)-\ln{(n+1)}}-\frac{1}{2n-\ln{n}}=\frac{\ln{(1+\frac{1}{n})}}{(2n-\ln{n})(2n+2-\ln{(n+1)})}<0$

так как

$\ln{(1+\frac{1}{n})}<\frac{1}{n}$ для всех $n$.

Следовательно, по признаку Лейбница данный ряд сходится.

Ответ: ряд $\sum_{n=1}^{\infty }\frac{(-1)^{n+1}}{2n-\ln{n}}$ сходится.


Исследовать ряд на сходимость:


$\sum_{n=1}^{\infty }\frac{(x-2)^{n+1}}{3^n(n+2)}$

РЕШЕНИЕ:

Для исследования сходимости данного ряда воспользуемся признаком Даламбера.

Пусть $a_n = \frac{(x-2)^{n+1}}{3^n(n+2)}$.

Тогда

$\frac{a_{n+1}}{a_n}=\frac{(x-2)^{n+2}}{3^{n+1}(n+3)}\cdot\frac{3^n(n+2)}{(x-2)^{n+1}}=\frac{(x-2)}{3}\cdot\frac{n+2}{n+3}.$ 

Для исследования сходимости данного ряда рассмотрим:

$x=2$. В этом случае $a_n=0$ для всех $n$, следовательно, ряд сходится.

$x\neq 2$. В этом случае

$\lim_{n\rightarrow \infty }\frac{a_{n+1}}{a_n}=\frac{(x-2)}{3}<1$


следовательно, ряд сходится по признаку Даламбера.

Ответ: ряд $\sum_{n=1}^{\infty }\frac{(x-2)^{n+1}}{3^n(n+2)}$ сходится при $x=2$ и при $\left|x-2\right|<3$.
