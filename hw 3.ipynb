#### Задание №1
Пусть $x_n = n^{(−1)^n}$

Доказать, что последовательность ${x_n}$:<br>
а) неограниченная;<br>
б) не является бесконечно большой.

##### Решение
##### Докажем, что последовательность неограничена.

Последовательность {$x_n$} называется неограниченной, если для любого сколь угодно большого M>0 найдётся такой номер $N_M$, что для любого n >= N |$y_n$| > M


Найдём номер $N_M$ члена последовательности, который первым окажется больше М = 100<br>
$$|n^{(-1)^n}| > 100 => n > 100$$
$$N_m = 101$$
т.е. все |$x_n$| начиная с 101 будут больше M
следовательно последовательность неограничена


##### Докажем, что последовательность не является бесконечно большой
Найдём предел функции $\lim\limits_{n\to\infty} n^{(-1)^n}$
Подставим $n = \infty$
$$\infty^{(-1)^\infty} = \infty$$

Предел функции равен бесконечности, значит последовательность **является бесконечно большой**

```python
def f(x):
    return x**((-1)**x)

x = np.linspace(1,100,100)
plt.plot(x,f(x))
```

#### Задание №2
Доказать, что последовательность {sin n} расходится.

##### Решение

Докажем от противного, пусть последовательность sin(n) является фундаментальной, в таком случае верно утверждение
$$\lim\limits_{n\to\infty} (\sin(n+1) - \sin(n-1)) = 0$$
т.к. $\sin(a) - \sin(b) = 2\cos(\frac{a+b}{2})sin(\frac{a-b}{2})$, получаем
$$\lim\limits_{n\to\infty} 2\cos(\frac{n+1+n-1}{2})\sin(\frac{n+1-(n-1)}{2}) = 0$$
$$\lim\limits_{n\to\infty} 2\cdot \cos(n)\cdot \sin(1) = 0 $$
для того чтоб данные равенство выполнялось, необходимо чтобы 
$$\lim\limits_{n\to\infty} \cos(n) = 0$$
но и $$\lim\limits_{n\to\infty} \sin(n) = 0$$, так же должно быть верно.
А из выражения $$sin^2(n)+cos^2(n) = 1$$ получается что предел sin должен быть равен 1.
Исходя из данного противоречия можно сделать вывод что последовательность sin(n) **расходится**

#### Задание №3
#### Найти пределы:
a) $\lim\limits_{n\to\infty} \frac{10n}{n^2+1}$
б) $\lim\limits_{n\to\infty} \frac{n^2-n}{n-\sqrt{n}}$
в) $\lim\limits_{n\to\infty} \frac{5 \cdot 3^n}{3^n -2 }$

##### Решение
а) $\lim\limits_{n\to\infty} \frac{10n}{n^2+1}$
$$\lim\limits_{n\to\infty} \frac{10n}{n^2+1} = {\frac{\infty}{\infty}} = \lim\limits_{n\to\infty} \frac{10\infty}{\infty^2+1} = \lim\limits_{n\to\infty} \frac{\infty}{\infty} = 0 $$

б) $\lim\limits_{n\to\infty} \frac{n^2-n}{n-\sqrt{n}}$
$$\lim\limits_{n\to\infty} \frac{n^2-n}{n-\sqrt{n}} = \frac{\infty}{\infty} = \lim\limits_{n\to\infty} \frac{\infty^2-\infty}{\infty-\sqrt{\infty}} = \lim\limits_{n\to\infty} \frac{\infty-\infty}{\infty-\infty} = \infty$$

в) $\lim\limits_{n\to\infty} \frac{5 \cdot 3^n}{3^n -2 }$
Разделим числитель и знаменатель на $3^n$
$$\lim\limits_{n\to\infty} \frac{5}{1 - \frac{2}{3^n}}$$
исходя из того что $\frac{1}{x^n} ⟶ 0$ при n>0 и $x ⟶ \infty$ получим
$$ \lim\limits_{n\to\infty} \frac{5}{1 - 0} = \lim\limits_{n\to\infty} 5 = 5$$

#### Задание №4
Найти предел
$\lim\limits_{n\to\infty} (\sqrt{n^2+n} - n)$

##### Решение
$$\lim\limits_{n\to\infty} (\sqrt{n^2+n} - n)$$
Умножим и разделим на сопряженное выражение $\sqrt{n^2+n} + n$ 
$$\frac{(\sqrt{n^2+n} - n)(\sqrt{n^2+n} + n)}{\sqrt{n^2+n} + n} = \frac{(\sqrt{n^2+n})^2 + n\cdot \sqrt{n^2+n} -n\cdot \sqrt{n^2+n} - n^2}{\sqrt{n^2+n} + n} = \frac{n}{\sqrt{n^2+n} + n}$$
разделим числитель и знаменатель на n
$$\frac{\frac{n}{n}}{\sqrt{n^2+n}/n + 1} = \frac{1}{\sqrt{1+\frac{1}{n}}+1}$$
исходя из того что $\frac{1}{n} ⟶ 0$ при $n ⟶ \infty$ получим
$$\lim\limits_{n\to\infty} \frac{1}{\sqrt{1+0}+1} = \frac{1}{2}$$

$$\lim\limits_{n\to\infty} \frac{\sqrt{n}\cdot \cos{n}}{n+1} = 0$$
